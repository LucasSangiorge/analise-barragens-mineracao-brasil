```python
import pandas as pd
```

```python
# Carregar dataset bruto
df = pd.read_csv(
    r"C:\Users\lucas\OneDrive\analise-barragens-mineracao-brasil\dados_brutos\Barragens.csv",
    encoding="latin1",
    engine="python"
)
```

```python
#visualização da estrutura
print(df.head())
print(df.columns)
```

```python
# Remove linhas totalmente vazias
df = df.dropna(how = "all")
```

```python
#padroniza nomes das colunas
df.columns = df.columns.str.lower().str.replace("","_")
```

```python
#Remove duplicados
df= df.drop_duplicates()
```

```python
#salvar dataset tratado
df.to_csv(r"C:\Users\lucas\OneDrive\analise-barragens-mineracao-brasil\dados_brutos\Barragens.csv", index = False)
```
