```python
import pandas as pd
import unicodedata
```

```python
# Carregar dataset bruto
df = pd.read_csv(
    r"C:\Users\lucas\OneDrive\analise-barragens-mineracao-brasil\dados_brutos\Barragens_tratado.csv",
    encoding="utf-8",
    sep=None,
    engine="python"
)

print(df.head())
```

```python
# Remove linhas totalmente vazias
df = df.dropna(how = "all")
```

```python
#padroniza nomes das colunas
df.columns = df.columns.str.lower().str.replace("","_")
```

```python
#Remove duplicados
df= df.drop_duplicates()
```

```python
df.columns = (
    df.columns
    .str.replace("_", "", regex=False)  # remove _
    .str.strip()                        # remove espaços
    .str.lower()                        # minúsculo
    .str.replace(" ", "_")              # espaço vira _
)
```

```python
import unicodedata

def remover_acentos(texto):
    return unicodedata.normalize('NFKD', texto).encode('ascii', 'ignore').decode('utf-8')

df.columns = [remover_acentos(col) for col in df.columns]
```

```python
colunas = [
    "id",
    "nome",
    "empreendedor",
    "uf",
    "municipio",
    "latitude",
    "longitude",
    "categoria_de_risco_-_cri",
    "dano_potencial_associado_-_dpa",
    "metodo_construtivo_da_back_up_dam"
]

df = df[colunas]
```

```python
import re

def dms_para_decimal(coord):
    if pd.isna(coord):
        return None
    
    numeros = re.findall(r'\d+\.?\d*', coord)
    
    if len(numeros) >= 3:
        graus = float(numeros[0])
        minutos = float(numeros[1])
        segundos = float(numeros[2])
        
        decimal = graus + minutos/60 + segundos/3600
        
        if "-" in coord:
            decimal = -decimal
            
        return decimal
```

```python
df["latitude"] = df["latitude"].apply(dms_para_decimal)
df["longitude"] = df["longitude"].apply(dms_para_decimal)
df.info()
```

```python
#salvar dataset tratado
df.to_csv(r"C:\Users\lucas\OneDrive\analise-barragens-mineracao-brasil\dados_brutos\Barragens_tratado_gold.csv", index = False)
```
